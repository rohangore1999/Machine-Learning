```python
import numpy as np
import pandas as pd
import sqlite3    ## SQL Interface
import pickle     ## Used to save your data - Converts objects to byte stream and vice versa
from sklearn.feature_extraction.text import CountVectorizer  ## BOW Model
```

```python
conn = sqlite3.connect('finalSummary.sqlite')

final = pd.read_sql_query("""SELECT * FROM Reviews""", conn)

final.head()
```

```python
def polarity(x):
    if x == 'positive':
        return 1
    return 0

final['Score']=final['Score'].map(polarity)
final.head()
```

```python
final.drop(['level_0'],axis=1,inplace=True)
```

```python
final.head()
```

```python
bow_vect = CountVectorizer()
bow = bow_vect.fit_transform(final['CleanedSummary'].values)
bow.shape
```

```python
final['Score'].iloc[900:10000].value_counts()
```

```python
X = bow[900:10000, :].toarray()
type(X)
```

```python
df = final[['Time','Score']]
df = df.iloc[900:10000]
df.head(20)
```

```python
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df1=pd.DataFrame(X)
df1.head(10)
```

```python
df=pd.concat([df,df1],axis=1)
df.head(10)
```

```python
df.sort_values('Time', inplace = True)
df.head()
```

```python
x=df.iloc[:,2:].values #removing score and time columns
y=df.iloc[:,1].values #just selecting Score& Time was just needed to sort
```

```python
type(x)
```

```python
type(y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score
```

```python
s = StandardScaler()
x=s.fit_transform(x)
```

```python
x_train = x[:700,:]
x_test = x[700:,:]
y_train = y[:700]
y_test = y[700:]
```

```python
y_test.shape
```

```python
from sklearn.cross_validation import cross_val_score
```

```python
# creating odd list of K for KNN
neighbors = list(range(1,20,2))
# empty list that will hold cv scores
cv_scores = []

# perform 10-fold cross validation
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())

# changing to misclassification error
MSE = [1 - x for x in cv_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print('\nThe optimal number of neighbors is %d.' % optimal_k)
```

```python
classifier = KNeighborsClassifier(n_neighbors = optimal_k)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred, normalize=True) * float(100)
acc
```
